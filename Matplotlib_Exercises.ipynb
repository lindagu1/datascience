```python
# YOUR CODE HERE

df.describe()

plt.figure()
plt.subplot(1,2,1)
plt.hist(df['Chol'], bins=30)
plt.title("Cholesterol before Norm")

def normalize(x):
    return (x - x.mean())/x.std()

df['Chol_norm'] = normalize(df['Chol'])
plt.subplot(1,2,2)
plt.hist(df['Chol_norm'], bins = 30)
plt.title("Cholesterol after Norm")


```

Conduct a Bivariate Analysis for `"Protein"` and `"Chol"`. What is familar about this compared to the tutorial? What is another way to deal with overplotting other than the Hex plot? Create two data visualizations, one that shows the overplotting, another that resolves this issue. After you are done, play around with the sns.kdeplot function or try to create a 3D graph to reduce the effects of overplotting.

```python
# YOUR CODE HERE
plt.figure()

plt.subplot(1,3,1)
plt.scatter(df['Protein'], df['Chol'])
plt.xlabel('Protein')
plt.ylabel('Chol')
plt.title('Scatter (Overplotting)')

plt.subplot(1,3,2)
plt.hexbin(df['Protein'], df['Chol'], gridsize=25)
plt.xlabel('Protein')
plt.ylabel('Chol')
plt.title('Hexbin (Reduced Overplotting)')
plt.colorbar()

plt.subplot(1,3,3)
sns.kdeplot(data=df, x='Protein', y='Chol', fill=True)
plt.xlabel('Protein')
plt.ylabel('Chol')
plt.title('KDE Plot')

plt.show()
```

Fill in the function below. The function recieves a non-empty string and an int n, and returns a new string where the character at index n has been removed.¶


```python
def remove_char(old_string, n):
    new_string = old_string[:n] + old_string[n+1:]
    return new_string

remove_char("words", 0)
"words"[len("words")-1:]
```

Given a string, return a new string where the first and last characters have been exchanged.

```python
def swap_chars(old_string):
  new_string = old_string[-1] + old_string[1:-1] + old_string[0]
  return new_string
```

Test the function with three different parameters


```python
# YOUR CODE HERE
print(swap_chars("hello"))
print(swap_chars("KTP"))
print(swap_chars("bad morning"))
```
